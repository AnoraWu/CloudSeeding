```python
from statsmodels.stats.power import TTestIndPower

n1 = 26
n2 = 25
power_analysis = TTestIndPower()

power_t_test_paper = power_analysis.power(effect_size=0.1277, nobs1=n1, alpha=0.05, ratio=n2/n1, alternative='larger')

power_t_test_paper
```

```python
sample_size = power_analysis.solve_power(effect_size=0.1277, nobs1=None, alpha=0.05, power=0.8, ratio=1.0, alternative='larger')

sample_size
```
