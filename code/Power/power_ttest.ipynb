```python
import numpy as np
from scipy.stats import nct, t

def power_from_final_framework(beta, s_x, n_w, n_v, alpha=0.05):
    """
    Computes the power of a one-sided two-sample t-test under the assumptions:
    - Equal variance across treatment and control groups
    - Effect size is d = beta / s_x
    - Power is computed using a one-sided test (H1: v̄ > w̄)
    
    Parameters:
    - beta: effect size estimate from your original study (ȳ - x̄)
    - s_x: standard deviation of the control group from your original study
    - n_w: sample size in the control group (unseeded group) in the new study
    - n_v: sample size in the treatment group (seeded group) in the new study
    - alpha: significance level (default = 0.05)
    
    Returns:
    - power: estimated power of the test
    """
    # Step 1: Compute standardized effect size
    d = beta / s_x

    # Step 2: Compute noncentrality parameter
    lambda_ncp = d / np.sqrt(1/n_w + 1/n_v)

    # Step 3: Degrees of freedom under equal variance assumption
    df = n_w + n_v - 2

    # Step 4: Critical t-value for one-sided test
    t_crit = t.ppf(1 - alpha, df)

    # Step 5: Power using noncentral t-distribution
    power = 1 - nct.cdf(t_crit, df, lambda_ncp)

    return power

# Example usage
beta = 0.6235  # Example: estimated from the PSM study, use coefficient of day 0.
s_x = 5.5687   # Standard deviation of control group from your study, s_x = 5.5687 for precipitation in day 0, 
n_w = 95     # Sample size in unseeded group in new study
n_v = 95  # Sample size in seeded group in new study

power = power_from_final_framework(beta, s_x, n_w, n_v, alpha=0.05)
print(f"Estimated power: {power:.4f}")

```
